```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('covid_data.csv')
```

```python
pd.reset_option('display.max_rows')
```

```python
data
```

```python
df = data.melt( id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'], var_name = 'date', value_name = 'deaths')
```

```python
df.duplicated().sum()
```

```python
df['previous_day_death'] = df.groupby('Country/Region')['deaths'].shift(1)
```

```python
pd.reset_option('display.max_rows')
```

```python
df[df['Country/Region'] == 'Nepal']
```

```python
df['Dailydeaths'] = df['deaths'] - df['previous_day_death']
```

```python
pd.reset_option('display.max_rows')
```

```python
df[df['Country/Region'] == 'Nepal']
```

```python
df.isnull().sum()
```

```python
df['Province/State'] = df['Province/State'].fillna('NotGiven')
```

```python
df['Lat'] = df['Lat'].fillna('NotGiven')
```

```python
df['Long'] = df['Long'].fillna('NotGiven')
```

```python
df.isnull().sum()
```

```python
df['previous_day_death'] = df['previous_day_death'].fillna(0)
```

```python
df['Dailydeaths'] = df['Dailydeaths'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['previous_day_death'] = df['previous_day_death'].astype(int)
```

```python
df['Dailydeaths'] = df['Dailydeaths'].astype(int)
```

```python
df.info()
```

```python
cleaned_dataset = df.to_csv('Cleaned_covid_data.csv')
```
